```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.info()
```

# 1.Line PLot

```python
plt.plot(iris['sepal_length'])
plt.title('Line PLot')
plt.xlabel('index')
plt.ylabel('sepal length')
plt.show()
```

# 2.Scatter plot

```python
plt.scatter(iris['sepal_length'],iris['sepal_width'])
plt.xlabel('Sepal length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Width')
plt.show()
```

# 3.Histogram

```python
plt.hist(iris['petal_length'], bins = 20)
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.title("histogram")
plt.show()
```

# 4. Box Plot

```python
plt.boxplot(iris['sepal_length'])
plt.title("Box Plot")
plt.show()
```

5.Scatter PLot with hue

```python
sns.scatterplot(
    x='sepal_length',
    y='sepal_width',
    hue='species',
    data=iris
)
plt.show()
```

6.PairPlot

```python
sns.pairplot(iris, hue='species')
```

Histogram

```python
sns.histplot(iris['petal_length'], kde=True)
sns.kdeplot(iris)
```

violin plot

```python
sns.violinplot(
    x='species',
    y='sepal_length',
    data=iris
)
plt.show()
```

count plot

```python
sns.countplot(
    x='species',
    data=iris
)
plt.show()
```

heatmap

```python
corr = iris.drop('species', axis=1).corr()

sns.heatmap(corr, annot=True)
plt.show()
```

```python
sns.boxplot(x='species',y='sepal_length',data=iris)
```

```python
sns.displot(iris)
```

```python
sns.regplot(x='sepal_length',y='petal_length',data=iris)
```

```python
sns.ecdfplot(
    data=iris,
    x="petal_width",
    hue="species"
)
```

```python
#jointplot
sns.jointplot(data=iris,x="sepal_length",y="sepal_width",hue="species")
```
